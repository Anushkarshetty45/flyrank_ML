### Quick Look at the Data

To understand whether the Content Refresh / Content Opportunity Scoring lane is useful, I explored the starter dataset.

I wanted to know:
- How many pages are available for analysis.
- How many pages are old enough to potentially require a content refresh.
- How many pages are still receiving significant search impressions.

These numbers help show whether there are enough pages that could benefit from prioritization instead of manually reviewing every page.

```python
import pandas as pd

df = pd.read_csv("data/raw/content_refresh_anonymized.csv")

print("Total Pages:", len(df))

old_pages = (df["days_since_last_update"] >= 180).sum()
print("Pages older than 180 days:", old_pages)

high_impression = (df["impressions_90d"] >= 500).sum()
print("Pages with more than 500 impressions:", high_impression)

print("Average CTR:", round(df["ctr"].mean(), 3))
```

The dataset contains thousands of webpages, which shows that manually checking every page would be time-consuming.

Many pages have not been updated for more than 180 days, while a significant number of pages are still receiving search impressions. This suggests that some older pages continue to attract visitors and may benefit from being refreshed.

These observations support my choice of the Content Refresh lane because prioritizing these pages can help focus effort where it is likely to have the greatest impact.
