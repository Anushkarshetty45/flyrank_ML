## 1. My lane as an ML task (type)

*Classification, clustering, ranking, or scoring — which one, and why?*

## My Lane as an ML Task

I have chosen the **Content Refresh / Content Opportunity Scoring** lane.

"This is primarily a ranking machine learning problem."

The goal is to assign each webpage a refresh priority score so that content teams know which pages should be updated first. Rather than making a simple yes/no decision, the model ranks webpages based on how likely they are to benefit from a content refresh.

The ranking is generated using multiple signals such as impressions, click-through rate (CTR), search position, and page age.

## 2. Target or proxy

*What would you predict? Where does that label come from — observed outcome or a defined rule?*

## Target or Proxy

The dataset does not contain a direct label indicating whether a webpage needs refreshing.

Instead, I will create a **proxy target** based on existing information such as:

- Page age
- Search impressions
- Click-through rate (CTR)
- Search position

Pages that are older, continue receiving impressions, but show weaker engagement are more likely to receive a higher refresh priority score,so whichever page has a higher refresh priority score ,that page will be refreshed.

## 3. Success metric

*One metric you can defend. What number means 'good'?*

## Success Metric

A successful model should rank webpages so that the pages most likely to benefit from a content refresh consistently appear near the top of the list. The ranking should help content teams prioritize updates more effectively than a simple rule based approach. The quality of the ranking can later be evaluated using ranking metrics such as NDCG or Precision@K once ground truth labels become available.

## 4. The unit of analysis, as a real dataframe

*Load your lane's slice and show it: one row = one what?*

## Unit of Analysis

Each row in the dataset represents **one webpage**.

Every webpage contains features such as search impressions, click-through rate, search position, page age, and other performance indicators.

These features describe the current state of a single webpage and will later be used to estimate its refresh priority.

```python
import pandas as pd

# Load the starter dataset
df = pd.read_csv("./data/raw/content_refresh_anonymized.csv")

# Display the first five rows
df.head()
```

```python
# Check the dataframe shape
print("Shape:", df.shape)

# Display important columns
df[['content_id',
    'impressions_90d',
    'clicks_90d',
    'ctr',
    'avg_position',
    'content_age_days',
    'trend_direction']].head()
```

```python
# simple example target
# Example proxy target for demonstration purposes only.
# This is NOT the final ML label.
df["refresh_priority"] = (
    (df["content_age_days"] > 180) &
    (df["ctr"] < 1.0)
)

df[["content_id",
    "content_age_days",
    "ctr",
    "refresh_priority"]].head(10)
```

## 5. Why ML beats a fixed rule here

*What makes the pattern too messy for an if-statement?*

## Why ML Beats a Fixed Rule

A fixed rule such as "refresh every page older than 180 days" is too simplistic because page performance depends on multiple factors.

Some older pages continue to perform well, while some newer pages may already be losing visibility.

Machine learning can evaluate several features together, including impressions, CTR, search position, and page age, to identify patterns that are difficult to capture using simple if-else rules.

This produces more useful recommendations and helps content teams prioritize updates more effectively.
